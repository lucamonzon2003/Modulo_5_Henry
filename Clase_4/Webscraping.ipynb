## Ejemplo 1: Ejercicio de práctica

Extraer los datos relativos a países productores de café, publicados en la página https://es.wikipedia.org/wiki/Anexo:Pa%C3%ADses_por_producci%C3%B3n_de_caf%C3%A9

```python
#Import needed libraries
from bs4 import BeautifulSoup as bs
import requests as rq
import pandas as pd
```

```python
url="https://en.wikipedia.org/wiki/List_of_countries_by_coffee_production"
page=rq.get(url).text # obtiene la página web
soup=bs(page)
soup
```

```python
table=soup.find('table') # encuentra la tabla en la página web 

#creamos un DataFrame vacío
df=pd.DataFrame(columns = ['country','bags_60kg','tons_metric','pounds']) 
df
```

```python
for row in table.find_all('tr')[1::]: #iteramos sobre cada fila de la tabla (excepto la primera) 
    cols = row.find_all("td") #obtenemos las columnas de cada fila 

cols
```

```python
#iterar sobre cada fila ('tr')
for row in table.find_all('tr')[1::]: # iteramos sobre cada fila de la tabla (excepto la primera) 
    cols = row.find_all("td") # obtenemos las columnas de cada fila 
    #print(cols)
    cols = [col.text.strip() for col in cols] # obtenemos el texto de cada columna y eliminamos los espacios en blanco 
    #print(cols)
    country=cols[1] #obtenemos el nombre del país 
    bags=cols[2]    #obtenemos el número de bolsas de 60kg
    tons=cols[3]   #obtenemos el número de toneladas métricas
    pounds=cols[4] #obtenemos el número de libras
    df=df.append({'country': country, 'bags_60kg': bags, 'tons_metric': tons,'pounds':pounds}, ignore_index=True)
```

## Ejemplo 2: Información de productos en una tienda

En este Notebook completaremos el ejercicio de extracción de datos desde la página web de una tienda de libros en línea. En el Tutorial, vimos como obtener la URL de la la página de cada producto, y hacer scarping de cada una de ellas para extrar los datos de nombre, precio, categoría, etc.

Sin embargo, al realizar el scraping de la página principal de la página, obtenemos un listado de sólo 20 productos, mientras que la tienda tiene 1000 libros.

Los resultados están dispuestos de a 20 libros por página. Es decir, para obtener la información de todos los productos a la venta, es necesario repetir el proceso para las 50 páginas de resultados de la tienda.
Eso haremos acá.

Recordemos el código anterior, que nos permite extraer los datos de los primeros 20 productos:

```python
#resumimos los pasos para llegar a la url de cada producto en una función
def getBooksURLs(url):
    result = requests.get(url)
    soup = BeautifulSoup(result.text, 'html.parser')

    lista=soup.findAll("article", class_ = "product_pod")
    book_url_list=['http://books.toscrape.com/catalogue/'+'/'+x.div.a.get('href') for x in lista]
    return(book_url_list)
```

¿Cómo recorremos las demás páginas de resultados? Note que al recorrer las páginas de resultados, la URL principal de la página va cambiando:

http://books.toscrape.com/catalogue/page-2.html <br>
http://books.toscrape.com/catalogue/page-3.html <br>
http://books.toscrape.com/catalogue/page-4.html <br>
http://books.toscrape.com/catalogue/page-5.html <br>

... y así sucesivamente hasta la página 50.

Por lo tanto, simplemente necesitamos repetir el proceso anterior, pero para las 50 páginas de resultados, para lo cual iremos cambiando el parámetro `main_url`.

```python
#para obtener las urls a todos los productos de la tienda, tenemos que recorrer varias páginas

#primero creamos la lista de páginas a recorrer
pages_urls = []

#para las páginas de 1 a 50
for i in range(1,51,1):
    #generamos la URL de cada página
    new_page="http://books.toscrape.com/catalogue/page-%d.html"%i # %d es un placeholder para el número de página (i)

    #agregamos un chequeo para asegurarmos de que la página está OK, si no la excluimos de la lista.
    if requests.get(new_page).status_code == 200: 
        pages_urls.append(new_page)

#Veamos algunos ejemplos
print("Ejemplos:")
pages_urls[:5]
```

```python
#ahora recorremos todas las páginas, para recolectar las url de todos los (1000) productos de la tienda
#la ejecución de esta celda puede demorar un poco más de lo habitual

booksURLs = []
for page in pages_urls:
    booksURLs.extend(getBooksURLs(page)) # extend() agrega los elementos de una lista (o cualquier iterable) al final de la lista actual (en este caso, booksURLs) y devuelve la lista extendida. 

print(str(len(booksURLs)) + " productos con URLs") #imprimimos la cantidad de productos encontrados (1000) y la palabra "productos con URLs" 
print("Ejemplos:")
booksURLs[::5]
```

Y finalmente repetimos el proceso de scraping para las 1000 páginas de productos:


```python
#la ejecución de esta celda puede demorar un poco más de lo habitual

names = []
prices = []
nb_in_stock = []
categories = []

for url in booksURLs:
    html = requests.get(url) # obtiene el contenido de la url y lo guarda en la variable html  
    soup = BeautifulSoup(html.text, 'html.parser') # crea un objeto BeautifulSoup a partir del contenido de la variable html y lo guarda en la variable soup 

    # Nombre del producto
    names.append(soup.find("div", class_ = "col-sm-6 product_main").h1.text) # obtiene el nombre del producto y lo guarda en la variable names  

    # Precio
    prices.append(soup.find("p", class_ = "price_color")) # obtiene el precio del producto y lo guarda en la variable prices 

    # Disponibilidad
    nb_in_stock.append(soup.find("p", class_ = "instock availability").text) # gobtiene la disponibilidad del producto y lo guarda en la variable nb_in_stock 

    # Categoria
    lines=soup.find("ul",class_='breadcrumb').find_all('li') # obtiene la categoría del producto y lo guarda en la variable categories 
    category=lines[2].a.text # obtiene la categoría del producto y lo guarda en la variable categories 
    categories.append(category) # obtiene la categoría del producto y lo guarda en la variable categories 

scraped_data = pd.DataFrame({'name': names, 'price': prices, 'nb_in_stock': nb_in_stock, "product_category": categories})
scraped_data
```

```python
#solo tomar valor numero de price con regex
scraped_data['price'] = scraped_data.price.apply(lambda x: float(x.text[2:])) # obtenemos el valor numérico del precio y lo convertimos a float para poder operar con él más adelante
```

```python
#eliminar caracteres no numericos de nb_in_stock

scraped_data['nb_in_stock'] = scraped_data.nb_in_stock.apply(lambda x: int(''.join(filter(str.isdigit, x)))) # obtenemos el valor numérico de la disponibilidad y lo convertimos a int para poder operar con él más adelante
```
